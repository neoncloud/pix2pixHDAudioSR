```python
b = 10
h = 5
x = 6
y = 7
d = 4
r = 8
q = torch.randn((b,h,x,y,d))
rel_k = torch.randn((r,d))
```

```python
e_logits = torch.einsum('b h x y d, r d -> b h x y r', q, rel_k)
logits = q.matmul(rel_k.transpose(-1, -2))
```

```python
print(logits.shape)
print(e_logits.shape)
print(logits-e_logits)
```
